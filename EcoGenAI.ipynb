```python
import gradio as gr
import numpy as np
import tensorflow as tf
from Bio.Seq import Seq
from Bio.SeqUtils import gc_fraction  # Correct function import
from Bio import SeqIO
import random

# Human Intelligence Model - Predict Life Threats without Harm
def human_ai_predict_threat(organism_name, genome_sequence):
    # Sample Intelligence: Detect harmful sequences based on pattern (hypothetical AI logic)
    threat_keywords = ["venom", "toxin", "poison", "aggression", "neurotoxin"]
    threat_score = sum(genome_sequence.lower().count(keyword) for keyword in threat_keywords)

    if threat_score == 0:
        return f"{organism_name} is SAFE. No harm detected."
    elif threat_score < 5:
        return f"{organism_name} has mild threat potential. Suggest observation."
    else:
        return f"{organism_name} may pose a HIGH THREAT. Avoid and protect biodiversity."

# Genome Sequencing Model for Harmful Species
def analyze_genome(genome_sequence):
    dna_seq = Seq(genome_sequence)
    gc_content = gc_fraction(dna_seq) * 100  # Fixed: Correct GC content calculation
    harmful_patterns = ["ATTG", "CCGG", "TATA", "GGCC"]  # Sample harmful patterns
    harmful_score = sum(genome_sequence.count(pattern) for pattern in harmful_patterns)

    if harmful_score > 10:
        return f"HIGH RISK species detected. Genetic analysis suggests harmful attributes!"
    elif harmful_score > 5:
        return f"MODERATE RISK species detected. Proceed with caution."
    else:
        return f"Species genome is SAFE. No harmful attributes detected."

# Gradio UI
def main():
    with gr.Blocks() as demo:
        gr.Markdown("## 🌍 AI-Powered Genome Sequencing & Human Intelligence Model")

        with gr.Tab("🔬 Analyze Harmful Species Genome"):
            genome_input = gr.Textbox(label="Enter Genome Sequence")
            genome_output = gr.Textbox(label="Analysis Result")
            analyze_btn = gr.Button("Analyze Genome")
            analyze_btn.click(analyze_genome, inputs=[genome_input], outputs=[genome_output])

        with gr.Tab("🧠 Human AI: Predict Threats"):
            organism_name = gr.Textbox(label="Organism Name")
            genome_sequence = gr.Textbox(label="Genome Sequence")
            prediction_output = gr.Textbox(label="AI Prediction")
            predict_btn = gr.Button("Predict Threat")
            predict_btn.click(human_ai_predict_threat, inputs=[organism_name, genome_sequence], outputs=[prediction_output])

    demo.launch()

# Run the UI
if __name__ == "__main__":
    main()
```

```python
import gradio as gr
from collections import Counter

# Pre-defined example datasets
example_datasets = {
    "Human Gene Fragment (Normal)": "ATGCGTAGCTAGCTAGCATCGATCGATCG",
    "Human Gene Fragment (Possible Mutation)": "ATGCGTGGCTAGCTAGCATCGATCGATCG",
    "Bacterial DNA Fragment": "GATTACAGATTACAGATTACAGATTA",
    "Viral DNA Fragment": "ACGUACGUACGUACGUACGUACGUACGU", # Note: Using U for RNA for demonstration
}

def analyze_sequence(dna_sequence):
    """Performs basic analysis: base counts and GC content."""
    dna_sequence = dna_sequence.upper().replace('U', 'T') # Treat U as T for simplicity in this basic example
    if not all(base in 'ATCG' for base in dna_sequence):
        return "Invalid DNA sequence. Please use only A, T, C, and G."
    base_counts = Counter(dna_sequence)
    gc_content = ((base_counts['G'] + base_counts['C']) / len(dna_sequence)) * 100 if len(dna_sequence) > 0 else 0
    results = f"**Base Counts:** A={base_counts['A']}, T={base_counts['T']}, C={base_counts['C']}, G={base_counts['G']}\n"
    results += f"**GC Content:** {gc_content:.2f}%"
    return results

def find_pattern(dna_sequence, pattern="GATA"):
    """Checks if a specific pattern exists in the DNA sequence using find() with debugging."""
    dna_sequence = dna_sequence.upper().replace('U', 'T')
    pattern = pattern.upper()
    print(f"Debugging: DNA Sequence being searched: '{dna_sequence}'")
    print(f"Debugging: Pattern being searched for: '{pattern}'")
    if dna_sequence.find(pattern) != -1:
        return f"The pattern '{pattern}' was not found in the sequence."
    else:
        return f"The pattern '{pattern}' was found in the sequence."

def full_analysis(dataset_name):
    """Performs both analyses on the selected dataset."""
    dna_sequence = example_datasets[dataset_name]
    basic_analysis_result = analyze_sequence(dna_sequence)
    pattern_finding_result = find_pattern(dna_sequence)
    return basic_analysis_result, pattern_finding_result

iface = gr.Interface(
    fn=full_analysis,
    inputs=gr.Dropdown(choices=list(example_datasets.keys()), label="Select a DNA Dataset"),
    outputs=[
        gr.Textbox(label="Basic Analysis (Base Counts, GC Content)"),
        gr.Textbox(label="Pattern Finding (Looking for 'GATA')")
    ],
    title="Simplified DNA Sequence Analyzer",
    description="Select a pre-defined DNA sequence to see basic analysis and check for a specific pattern."
)

iface.launch(share=True)
```
